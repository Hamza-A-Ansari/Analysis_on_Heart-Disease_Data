### START MACHINE LEARNING FROM HERE
### MODEL BUILDING

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
```

##### Here We are Using Train_Test_Split

```python
# First way

arr = data.values
x = arr[:,0:-1]
y = arr[:,-1]

testing_size = 0.2
print(y)
```

```python
#Second way

x = data.drop("target", axis=1)
y = data["target"]

print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size=testing_size)
```

```python
print(y_test)
```

#### Here First We Will Know The Accuracy Then Plot The Confusion Matrix By Using Different Models
### Our Goal To Achieve 90% to 100% Accuracy
### DECISION TREE CLASSIFIER()

```python
DT = DecisionTreeClassifier()
DT.fit(x_train, y_train)
y_pred = DT.predict(x_test)
acc = accuracy_score(y_test, y_pred)*100
print(f"Decision Tree accuracy score is: {acc}")

sns.set(font_scale=1.5)
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True, fmt="g")
plt.show()
```

### RANDOM FOREST CLASSIFIER()

```python
RT = RandomForestClassifier()
RT.fit(x_train, y_train)
y_pred = RT.predict(x_test)
acc = accuracy_score(y_test, y_pred)*100
print(f"Random Forest accuracy score is: {acc}")

sns.set(font_scale=1.5)
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True, fmt="g")
plt.show()
```

#### K NEIGHBORS CLASSSIFIER

```python
KNN =  KNeighborsClassifier()
KNN.fit(x_train, y_train)
y_pred = KNN.predict(x_test)
acc = accuracy_score(y_test, y_pred)*100
print(f"KNeighbors Classifier accuracy score is: {acc}")

sns.set(font_scale=1.5)
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True, fmt="g")
plt.show()
```

#### LOGISTIC REGRESSION

```python
LR =  LogisticRegression()
LR.fit(x_train, y_train)
y_pred = LR.predict(x_test)
acc = accuracy_score(y_test, y_pred)*100
print(f"LOGISTIC REGRESSION accuracy score is: {acc}")

sns.set(font_scale=1.5)
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True, fmt="g")
plt.show()
```

#### GAUSSIAN NB

```python
GNB = GaussianNB()
GNB.fit(x_train, y_train)
y_pred = GNB.predict(x_test)
acc = accuracy_score(y_test, y_pred)*100
print(f"GAUSSIAN NB accuracy score is: {acc}")

sns.set(font_scale=1.5)
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True, fmt="g")
plt.show()
```

# NOTE: ONLY DECISION TREE CLASSSIFIER AND RANDOM FOREST CLASSIFIER ARE BEST FOR THIS DATASET CAUSE
## THESE TWO MODELS ARE GIVING 95% TO 100% ACCURACY FOR THIS DATASET.
## KNEIGHBORS CLASSIFIER, GAUSSIAN NB, LOGISTIC REGRESSION THESE THREE MODELS ARE NOT BEST FOR THIS DATASET CAUSE
